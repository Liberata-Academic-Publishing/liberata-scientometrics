```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

def compute_h_index(citations):
    sorted_desc = sorted(citations, reverse=True)
    h = 0
    for rank, cites in enumerate(sorted_desc, start=1):
        if cites >= rank:
            h = rank
        else:
            break
    return h

# Generate 200 researchers, each with a random number of papers and Zipf-distributed citations
h_indices = []
for _ in range(200):
    n_papers = rng.integers(5, 60)
    citations = [min(int(c), 300) for c in rng.zipf(a=2.1, size=n_papers)]
    h_indices.append(compute_h_index(citations))

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(h_indices, bins=range(0, max(h_indices) + 2), color='#2563EB', edgecolor='white', linewidth=0.6)
ax.axvline(np.mean(h_indices), color='#DC2626', linewidth=1.8, linestyle='--', label=f'Mean = {np.mean(h_indices):.1f}')
ax.axvline(np.median(h_indices), color='#16A34A', linewidth=1.8, linestyle='--', label=f'Median = {np.median(h_indices):.0f}')
ax.set_xlabel('H-index', fontsize=12)
ax.set_ylabel('Number of researchers', fontsize=12)
ax.set_title('H-index distribution across 200 researchers', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()

print(f'Min: {min(h_indices)}  Max: {max(h_indices)}  Mean: {np.mean(h_indices):.1f}  Median: {np.median(h_indices):.0f}')
```
